```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 数据路径
dataset_path = r"D:\data\data03.csv"

# 加载数据
data = pd.read_csv(dataset_path)

# 提取前11列控制变量
control_variables = data.iloc[:, :11]

# 特征名修改
# 使用 LaTeX 格式来表示下标
control_variables.columns = [
    col.replace("Average_pore_size", "S$_{avg}$")  # 将“Average pore size”改为“S_avg”
    .replace("Total_pore_volume", "V$_{total}$")  # 将“Total pore volume”改为“V_total”
    .replace("ID_IG", "ID/IG")                    # 将“ID_IG”中的下划线改为“/”
    .replace("La_Lc", "La/Lc")                    # 将“La_Lc”中的下划线改为“/”
    for col in control_variables.columns
]

# 计算皮尔逊相关系数矩阵
correlation_matrix = control_variables.corr()

# 打印相关系数矩阵
print("皮尔逊相关系数矩阵：")
print(correlation_matrix)

# 可视化相关系数矩阵
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0,
            xticklabels=correlation_matrix.columns, yticklabels=correlation_matrix.columns)
plt.title("控制变量之间的皮尔逊相关系数矩阵")
plt.tight_layout()
plt.show()
# 保存图片
output_path = r"D:\data\correlation_matrix.png"  # 指定保存路径和文件名
plt.savefig(output_path, dpi=300, bbox_inches='tight')  # 保存为 PNG 格式，分辨率 300 DPI


print(f"热力图已保存到：{output_path}")
```
